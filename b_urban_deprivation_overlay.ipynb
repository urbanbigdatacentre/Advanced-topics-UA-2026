# Data preparation

The following lines of code read the datasets generated in the Sketchmap tool and the Kobo Toolbox survey. 

Note: You might require to change the path to the files depending on where you have stored them. It is preferred if you upload the files to google colab and read them from there.

```python
# File containing the results of digitisin the sketchmap tool
smt_path = Path("smt_path_to_fike.geojson")
smt_gdf = gpd.read_file(smt_path.as_posix())
smt_wgs84 = smt_gdf.to_crs(epsg=4326)

# File containing the results form the KoboToolbox survey
kobo_path = Path("smt_path_to_fike.geojson")
kobo_gdf = gpd.read_file(kobo_path.as_posix())
kobo_wgs84 = smt_gdf.to_crs(epsg=4326)
```

The following lines of code read the dataset containing the Scottish Index of Multiple Deprivation (SIMD). This dataset is an extraction of the original file available at https://www.gov.scot/collections/scottish-index-of-multiple-deprivation-2020/ and contains the SIMD deciles for the city of Glasgow. You can replace this dataset with any other reference dataset that you consider relevant for your analysis.


```python
raw_url = "https://raw.githubusercontent.com/urbanbigdatacentre/Advanced-topics-UA-2026/main/Glasgow_City_SIMD_2020.geojson"
    print(f"{geojson_path} not found locally. Downloading from: {raw_url}")
    urllib.request.urlretrieve(raw_url, geojson_path)

# Read the GeoJSON file and project to WGS84
simd = gpd.read_file(geojson_path.as_posix())
simd_wgs84 = simd.to_crs(epsg=4326)
```

```python
import folium

# Center map using SIMD polygons
center = [simd_wgs84.geometry.centroid.y.mean(), simd_wgs84.geometry.centroid.x.mean()]
m = folium.Map(location=center, zoom_start=11, tiles="OpenStreetMap")

# SIMD polygons with color legend based on CrimeCount
simd_wgs84.explore(
    m=m,
    name="SIMD (CrimeCount)",
    column="CrimeCount",
    cmap="YlOrRd",
    legend=True,
    tooltip=["CrimeCount"],
    style_kwds={"weight": 0.6, "color": "black", "fillOpacity": 0.6},
)

# Sketch map geometries
smt_wgs84.explore(
    m=m,
    name="Sketchmap",
    color="royalblue",
    style_kwds={"weight": 2},
)

# Kobo survey geometries
kobo_wgs84.explore(
    m=m,
    name="Kobo survey",
    color="seagreen",
    style_kwds={"weight": 2},
)

folium.LayerControl(collapsed=False).add_to(m)
m
```
